# **Tema del dataset**
El dataset elegido contiene informacin sobre partidos de ftbol argentino, incluyendo detalles como los equipos locales y visitantes, los resultados de los partidos y la fecha y el nombre de los torneos en los que se jugaron.

### **Definicin de las variables presentes en el dataset:**
date_name: La fecha y el nombre del partido o torneo.

local_team_id: El ID del equipo local.

local_result: El resultado obtenido por el equipo local en el partido.

visitor_result: El resultado obtenido por el equipo visitante en el partido.

### **date_name (Fecha y Nombre del Partido o Torneo):**
### *Tipo de Variable: Categrica.*
Esta variable representa la fecha y el nombre de un partido o torneo. Aunque contiene informacin de fecha, se utiliza ms como una etiqueta categrica que identifica el evento especfico. No se realizan operaciones aritmticas o de tiempo con esta variable.

### **local_team_id (ID del Equipo Local):**
### *Tipo de Variable: Numrica Discreta.*
Este es un identificador numrico nico para el equipo local en cada partido. Aunque es un nmero, se utiliza para identificar categricamente un equipo y no tiene sentido realizar operaciones aritmticas con l.

### **local_team (Nombre del Equipo Local):**
### *Tipo de Variable: Categrica.*
Esta variable contiene el nombre del equipo local en cada partido. Es una etiqueta categrica que identifica el equipo y no se utilizan operaciones matemticas con ella.

### **local_result (Resultado del Equipo Local):**
### *Tipo de Variable: Numrica Discreta.*
Esta variable representa el resultado obtenido por el equipo local en cada partido. Es un nmero discreto que indica la cantidad de goles anotados por el equipo local.

### **visitor_result (Resultado del Equipo Visitante):**
### *Tipo de Variable: Numrica Discreta.*
Similar a local_result, esta variable representa el resultado obtenido por el equipo visitante en cada partido y se expresa como un nmero discreto que indica la cantidad de goles anotados por el equipo visitante.

### **visitor_team (Nombre del Equipo Visitante):**
### *Tipo de Variable: Categrica.*
Al igual que local_team, esta variable contiene el nombre del equipo visitante en cada partido y se considera una etiqueta categrica.

### **visitor_team_id (ID del Equipo Visitante):**
### *Tipo de Variable: Numrica Discreta.*
Al igual que local_team_id, esta variable es un identificador numrico nico para el equipo visitante en cada partido y se usa principalmente para identificacin categrica.


```python
#########################################
#            EJERCICIO 4                #
#########################################
# Detectar Valores Ausente y Valores Atipicos .Decidir si eliminarlos y el por que de la eleccion.

# Valores Ausentes:
import pandas as pd

# Cargar el dataset
data = pd.read_csv('futbol-arg.csv')

# Identificar valores ausentes
valores_ausentes = data.isnull().sum()
print(valores_ausentes)

# Valores Atipicos:
import matplotlib.pyplot as plt

# Diagrama de caja para identificar valores atpicos
plt.figure(figsize=(12, 6))
data.boxplot()
plt.xticks(rotation=90)
plt.show()

```

## En el caso de este dataset no considero eliminar ningun valor puesto que para este caso cada dato es informacion precisa y necesaria para lo que se quiere mostrar

```python
#########################################
#            EJERCICIO 5                #
#########################################
# Realizar un analisis univariado y en base a esos graficos,sacar conclusiones.

# Se grafica algunas visualizaciones univariadas de las variables numricas del dataset, como local_result y visitor_result, que representan
# los resultados de los equipos locales y visitantes en trminos de goles anotados.
import matplotlib.pyplot as plt

# Histograma de resultados locales
plt.figure(figsize=(8, 6))
plt.hist(data['local_result'], bins=10, color='blue', alpha=0.7)
plt.title('Histograma de Resultados Locales')
plt.xlabel('Goles Anotados por el Equipo Local')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de resultados visitantes
plt.figure(figsize=(8, 6))
plt.hist(data['visitor_result'], bins=10, color='green', alpha=0.7)
plt.title('Histograma de Resultados Visitantes')
plt.xlabel('Goles Anotados por el Equipo Visitante')
plt.ylabel('Frecuencia')
plt.show()
```

# **Conclusiones:**

A partir del anlisis de los histogramas, se puede observar la distribucin de los resultados de los equipos locales y visitantes. Por ejemplo, si la mayora de los partidos tienen pocos goles o si hay una tendencia hacia puntajes ms altos.


Tambin puede realizarse un anlisis estadstico ms detallado, como calcular la media, la mediana y la desviacin estndar de los resultados locales y visitantes para una comprensin ms completa de la distribucin.

```python
#########################################
#            EJERCICIO 6                #
#########################################
# Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset (reemplaza 'nombre_del_archivo.csv' con el nombre real de tu archivo)
data = pd.read_csv('futbol-arg.csv')

# Calcular la matriz de correlacin
correlation_matrix = data.corr()

# Crear una figura para el grfico
plt.figure(figsize=(10, 8))

# Crear un mapa de calor (heatmap) de la matriz de correlacin
heatmap = plt.matshow(correlation_matrix, cmap='coolwarm')
plt.colorbar(heatmap)

# Etiquetar los ejes y mostrar el grfico
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Matriz de Correlacin')

# Mostrar el grfico
plt.show()
```

En el mapa de calor se puede identificar las variables que estn correlacionadas.

Si dos variables tienen un color ms claro en el mapa de calor, indican una correlacin positiva (cuando una variable aumenta, la otra tambin tiende a aumentar).

Si dos variables tienen un color ms oscuro, indican una correlacin negativa (cuando una variable aumenta, la otra tiende a disminuir).

Si una variable tiene una correlacin cercana a 1 con otra variable, estn altamente correlacionadas positivamente.

Si una variable tiene una correlacin cercana a -1 con otra variable, estn altamente correlacionadas negativamente.
