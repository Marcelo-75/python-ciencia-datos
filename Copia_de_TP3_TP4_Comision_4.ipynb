Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
###########################################################
#                     EJERCICIO 1.                        #
###########################################################
import numpy as np
import matplotlib.pyplot as plt

# Simulación de lanzamiento del dado
np.random.seed(0)  # Fijamos una semilla para reproducibilidad
num_lanzamientos = 10000  # Número de lanzamientos

lanzamientos = np.random.randint(1, 7, num_lanzamientos)  # Generar números aleatorios entre 1 y 6 (ambos incluidos)

# Calcular probabilidades aproximadas de obtener 2, 3 y 5
prob_2 = np.sum(lanzamientos == 2) / num_lanzamientos
prob_3 = np.sum(lanzamientos == 3) / num_lanzamientos
prob_5 = np.sum(lanzamientos == 5) / num_lanzamientos

# Imprimir las probabilidades
print(f"Probabilidad aproximada de obtener el número 2: {prob_2:.2f}")
print(f"Probabilidad aproximada de obtener el número 3: {prob_3:.2f}")
print(f"Probabilidad aproximada de obtener el número 5: {prob_5:.2f}")

# Graficar el histograma de las probabilidades
valores = [2, 3, 5]
probs = [prob_2, prob_3, prob_5]

plt.bar(valores, probs, tick_label=valores)
plt.xlabel('Número en el dado')
plt.ylabel('Probabilidad')
plt.title('Histograma de probabilidades')
plt.show()
```

Dada la base de datos de ventas diarias durante una semana, de un negocio con 4 productos.- VER EJERCICIO PARTE 2 PRACTICO N 2- Realizar el siguiente análisis de datos para poder obtener información y planificar una estrategia de ventas.

4.   Calcular el desvio estandar de los datos por cada producto y concluir para cada producto como es su rango de variabilidad en las ventas de los mismos.Plotear los datos de ventas de la semana de los productos con mayor y menor desvio estandar y evidenciar la variabilidad indicada en el parametro *'DESVIO ESTANDAR'*



```python
###########################################################
#                     EJERCICIO 4.1                       #
###########################################################
import numpy as np
import matplotlib.pyplot as plt

# Matriz de ventas diarias
ventas_diarias = np.array(
      #Lun,Mar,Mie,Jue,Vie,Sab,Dom
      [[20, 15, 25, 30, 18, 22, 24], #Producto A
      [12, 20, 14, 8, 15, 18, 16],  #Producto B
      [35, 28, 32, 30, 26, 24, 30], #Producto C
      [40, 38, 45, 42, 39, 41, 37]  #Producto D
      ]
)

# Calcular la media y la mediana por producto
media_por_producto = np.mean(ventas_diarias, axis=1)
mediana_por_producto = np.median(ventas_diarias, axis=1)

# Etiquetas de productos
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']

# Crear gráfico de barras
plt.figure(figsize=(10, 5))
plt.bar(productos, media_por_producto, label='Media')
plt.bar(productos, mediana_por_producto, label='Mediana', alpha=0.7)
plt.xlabel('Producto')
plt.ylabel('Ventas Promedio')
plt.title('Media y Mediana de Ventas por Producto')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
###########################################################
#                     EJERCICIO 4.2                       #
###########################################################
# Calcular la media y la mediana por día
media_por_dia = np.mean(ventas_diarias, axis=0)
mediana_por_dia = np.median(ventas_diarias, axis=0)

# Etiquetas de días
dias = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']

# Crear gráfico de trazo de líneas
plt.figure(figsize=(10, 5))
plt.plot(dias, media_por_dia, label='Media')
plt.plot(dias, mediana_por_dia, label='Mediana', linestyle='--', marker='o')
plt.xlabel('Día de la Semana')
plt.ylabel('Ventas Promedio')
plt.title('Media y Mediana de Ventas por Día')
plt.legend()
plt.grid(True)
plt.show()
```

* EDAD= entre 28 y 60 años

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.

```python
###########################################################
#                     EJERCICIO 5.0                       #
###########################################################
import random

# Solicitar al usuario la cantidad de datos a cargar
N = int(input("Ingrese la cantidad de datos a cargar: "))

# Definir las listas para almacenar los datos generados
sexos = ['MASCULINO', 'FEMENINO', 'NO ESPECIFICA']
datos_generados = []

# Generar datos aleatorios
for _ in range(N):
    sexo = random.choice(sexos)
    edad = random.randint(28, 60)
    altura = round(random.uniform(165.0, 210.0), 2)
    peso = random.randint(50, 110)
    horas_descanso = round(random.uniform(4.0, 12.0), 2)
    horas_trabajo_estudio = round(random.uniform(0.0, 16.0), 2)
    horas_ejercicios = round(random.uniform(0.0, 3.0), 2)

    # Agregar los datos generados a la lista
    datos_generados.append({
        'SEXO': sexo,
        'EDAD': edad,
        'ALTURA': altura,
        'PESO': peso,
        'HORAS DE DESCANSO diario': horas_descanso,
        'HORAS DE TRABAJO/ESTUDIO diario': horas_trabajo_estudio,
        'HORAS DE EJERCICIOS diario': horas_ejercicios
    })

# Imprimir los datos generados
print("\nDatos generados:")
for i, datos in enumerate(datos_generados, 1):
    print(f"\nPersona {i}:")
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")
```

```python
###########################################################
#                     EJERCICIO 5.1                       #
###########################################################
# Generar un array de dimensión 2 (matriz) con todos los datos ordenados para visualizarlos como tabla:

import pandas as pd

# Crear un DataFrame con los datos generados
df = pd.DataFrame(datos_generados)

# Mostrar el DataFrame
print("Tabla de Datos:")
print(df)
```

```python
###########################################################
#                     EJERCICIO 5.2                       #
###########################################################
# Encontrar la media en los datos de altura, peso y horas de descanso, trabajo y ejercicios:

# Calcular la media de altura, peso y horas
media_altura = df['ALTURA'].mean()
media_peso = df['PESO'].mean()
media_horas_descanso = df['HORAS DE DESCANSO diario'].mean()
media_horas_trabajo_estudio = df['HORAS DE TRABAJO/ESTUDIO diario'].mean()
media_horas_ejercicios = df['HORAS DE EJERCICIOS diario'].mean()

print(f"Media de Altura: {media_altura:.2f} mts")
print(f"Media de Peso: {media_peso:.2f} kg")
print(f"Media de Horas de Descanso: {media_horas_descanso:.2f} horas")
print(f"Media de Horas de Trabajo/Estudio: {media_horas_trabajo_estudio:.2f} horas")
print(f"Media de Horas de Ejercicios: {media_horas_ejercicios:.2f} horas")
```
